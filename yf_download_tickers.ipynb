```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Define URL for S&P 500 components page
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
# url = 'https://en.wikipedia.org/wiki/Nasdaq-100'

# Send GET request to URL and get HTML response
response = requests.get(url)

# Parse HTML content with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find table containing S&P 500 components
sp500_table = soup.find('table', {'class': 'wikitable sortable'})

# Extract tickers from table and store in list
sp500_tickers = []
for row in sp500_table.findAll('tr')[1:]:
    ticker = row.findAll('td')[0].text.strip()
    sp500_tickers.append(ticker)

# # Use yfinance to get information on S&P 500 tickers
# sp500_data = yf.download(sp500_tickers, period="1d")

# Print tickers and current prices
print(f'sp500 tickers len({len(sp500_tickers)}):\n{sp500_tickers}')
```

```python
# Define URL for S&P 500 components page
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_400_companies'

# Send GET request to URL and get HTML response
response = requests.get(url)

# Parse HTML content with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find table containing S&P 500 components
sp400_table = soup.find('table', {'class': 'wikitable sortable'})

# Extract tickers from table and store in list
sp400_tickers = []
for row in sp400_table.findAll('tr')[1:]:
    ticker = row.findAll('td')[0].text.strip()
    sp400_tickers.append(ticker)

print(f'sp400 tickers len({len(sp400_tickers)}):\n{sp400_tickers}')
```

```python
# Define URL for S&P 500 components page
url = 'https://en.wikipedia.org/wiki/Nasdaq-100'

# Send GET request to URL and get HTML response
response = requests.get(url)

# Parse HTML content with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find(lambda tag: tag.name=='table' and tag.has_attr('id') and tag['id']=="constituents") 
rows = table.findAll(lambda tag: tag.name=='tr')

# Extract tickers from table and store in list
nasdaq100_tickers = []
# for row in rows:
for row in rows[1:]:
    ticker= row.find_all('td')[1].text.strip()
    nasdaq100_tickers.append(ticker)

# # Print tickers and current prices
print(f'nasdaq100 tickers len({len(nasdaq100_tickers)}):\n{nasdaq100_tickers}')
```

```python
symbols_sp500_400_ndx_100 = sp500_tickers + sp400_tickers + nasdaq100_tickers
print(len(symbols_sp500_400_ndx_100))
```
